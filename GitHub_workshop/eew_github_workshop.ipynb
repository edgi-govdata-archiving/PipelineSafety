This Python notebook utilizes Hazardous liquid data to do some basic analysis and make a figure

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
gt = pd.read_csv("data/PHMSA_clean/gt_inc.csv")
```

```python
gt
```

```python
# Prepare data for plotting 
gt_sig = gt[gt['SIGNIFICANT'] == "YES"]
gt_sig = gt_sig[gt_sig['IYEAR'] < 2026]
cause_yr = gt_sig.groupby(['IYEAR', 'MAP_EIGHT_CAUSE']).size().reset_index(name = 'total')
```

```python
# Plot incident over time by cause
sns.relplot(data=cause_yr, x='IYEAR', y='total', hue='MAP_EIGHT_CAUSE',kind='line')
plt.title("Significant incidents by cause\n2010--2025")
```
